## 题目 1：PromptTemplate 基础

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template(
    "用{language}写一个{function_name}函数，要求：{requirement}"
)
print(prompt.input_variables)

value = prompt.invoke({
    "language": "Java",
    "function_name": "reverseString",
    "requirement": "原地反转，不使用额外空间",
})
print(value.to_string())
```

## 题目 2：ChatPromptTemplate 元组写法

```python
from langchain_core.prompts import ChatPromptTemplate

prompt2 = ChatPromptTemplate.from_messages([
    ("system", "你是一个{role}"),
    ("human", "请帮我{task}"),
])
msgs = prompt2.invoke({"role": "Java 专家", "task": "解释 HashMap 的扩容机制"})
for m in msgs.to_messages():
    print(f"[{m.__class__.__name__}] {m.content}")
```

## 题目 3：ChatPromptTemplate 类写法

```python
from langchain_core.prompts import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
)

prompt3 = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template("你是一个{role}"),
    HumanMessagePromptTemplate.from_template("请帮我{task}"),
])
msgs3 = prompt3.invoke({"role": "Java 专家", "task": "解释 HashMap 的扩容机制"})
for m in msgs3.to_messages():
    print(f"[{m.__class__.__name__}] {m.content}")
```

## 题目 4：MessagesPlaceholder

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage

prompt4 = ChatPromptTemplate.from_messages([
    ("system", "你是一个代码审查助手"),
    MessagesPlaceholder(variable_name="history"),
    ("human", "{input}"),
])

history = [
    HumanMessage(content="我写了一个冒泡排序，时间复杂度是 O(n²)"),
    AIMessage(content="冒泡排序确实是 O(n²)，数据量大时建议换快速排序。"),
]

filled = prompt4.invoke({"history": history, "input": "那快速排序平均复杂度是多少？"})
for m in filled.to_messages():
    print(f"[{m.__class__.__name__}] {m.content}")
```

## 题目 5：partial 预填

```python
from langchain_core.prompts import PromptTemplate

prompt5 = PromptTemplate.from_template("{language}中，{concept}是什么意思？")
partial_prompt = prompt5.partial(language="Java")

print(partial_prompt.input_variables)   # 只剩 ['concept']

value5 = partial_prompt.invoke({"concept": "泛型擦除"})
print(value5.to_string())
```
