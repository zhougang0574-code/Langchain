## 题目 1：PromptTemplate 基础

要求：
1. 用 `from_template` 创建一个模板：`"用{language}写一个{function_name}函数，要求：{requirement}"`
2. 打印它的 `input_variables`
3. 用 `invoke` 填入任意值，打印 `to_string()` 的结果

```python

from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template("用{language}写一个{function_name}函数，要求：{requirement}")
print(template.input_variables)
message = template.invoke({"language": "Python", "function_name": "递归", "requirement": "有中文注释"})
print(message.to_string())
```

## 题目 2：ChatPromptTemplate 元组写法

要求：
1. 用**元组写法**创建 `ChatPromptTemplate`：
   - system：`你是一个{role}`
   - human：`请帮我{task}`
2. `invoke` 填值后，用 `to_messages()` 遍历打印每条消息的角色类名和内容

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个{role}"),
    ("human", "请帮我{task}")
])

message = prompt.invoke({"role": "AI助手", "task": "做项目"})
for m in message.to_messages():
    print(f"[{m.__class__.__name__}] {m.content}")
```

## 题目 3：ChatPromptTemplate 类写法

要求：
1. 用 `SystemMessagePromptTemplate` 和 `HumanMessagePromptTemplate` 重写题目 2
2. 填相同的值，验证输出和元组写法完全一致

```python
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

prompt = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template("你是一个{role}"),
    HumanMessagePromptTemplate.from_template("请帮我{task}")
])

message = prompt.invoke({"role": "AI助手", "task": "做项目"})
print(message.to_string())
```

## 题目 4：MessagesPlaceholder

要求：
1. 创建带历史槽的 `ChatPromptTemplate`：
   - system：`你是一个代码审查助手`
   - history 槽
   - human：`{input}`
2. 手动构造一段历史（至少一问一答，用 `HumanMessage` 和 `AIMessage`）
3. `invoke` 填值后打印展开的完整消息列表，确认历史被正确插入

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder, HumanMessagePromptTemplate
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

template = ChatPromptTemplate.from_messages([
    SystemMessage(content="你是一个代码审查助手"),
    MessagesPlaceholder(variable_name="history"),
    HumanMessagePromptTemplate.from_template("{input}")
])

history = [
    HumanMessage(content="今天是星期二，你记住了"),
    AIMessage(content="好的，我记牢了")
]

message = template.invoke({"history": history, "input": "今天天气如何"})
print(message.to_string())
```

## 题目 5：partial 预填

要求：
1. 创建模板：`"{language}中，{concept}是什么意思？"`
2. 用 `partial` 把 `language` 预填为 `"Java"`
3. 打印预填后对象的 `input_variables`（应该只剩 `['concept']`）
4. 再用 `invoke` 只传 `concept`，打印最终文本

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template("{language}中，{concept}是什么意思？").partial(language="Java")
print(prompt.input_variables)
message = prompt.invoke({"concept": "递归"})
print(message.to_string())
```
